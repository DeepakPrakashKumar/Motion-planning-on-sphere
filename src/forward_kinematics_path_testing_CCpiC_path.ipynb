# Path checking for CCpiC path
We desire to verify the construction of the CCpiC path.

```python
import math
from Path_generation_sphere import *

ini_config = np.eye(3)

def function_testing_solution_CCpiC(r_array, phi1_array, phi3_array, path_type):

    for r in r_array:

        print('Considering r value of ', r)
        
        for phi1 in phi1_array:
            for phi3 in phi3_array:

                # Obtaining the final configuration
                fin_config = points_path(ini_config, r, 1, [phi1, math.pi, phi3], path_type)[3]
                
                # print('Final configuration is ', fin_config)

                path_params = path_generation_C_Cpi_C(ini_config, fin_config, r, 1, path_type)

                flag = 0
                # print(path_params)
                for path in path_params:

                    # We check if the turning radius is sqrt(2)
                    if abs(r - 1/math.sqrt(2)) > 10**(-6): # This is the same tolerance used in the code.

                        max_val = max([abs(path[1] - phi1), abs(path[-1] - phi3)])
                    
                    else:

                        max_val = max([abs(math.cos(path[1] - path[-1]) - math.cos(phi1 - phi3)),\
                                       abs(math.sin(path[1] - path[-1]) - math.sin(phi1 - phi3))])
                        # print('Max value is ', max_val)
                    
                    if max_val <= 10**(-4):

                        # print('Found correct path for ', phi1, phi2, phi3, ' angles for ', path_type, ' with parameters ', path)
                        flag = 1
                
                if flag == 0:
                    raise Exception('Found exception case for ', phi1, math.pi, phi3, ' angles for ', path_type, ' with parameters ', path)
    
    print('Passed the tests.')
    # return flag
```

```python
path_type_arr = ['lrl', 'rlr']

# r_array = [1/math.sqrt(2) - 10**(-10), 1/math.sqrt(2) - 10**(-9), 1/math.sqrt(2) - 10**(-8),\
#             1/math.sqrt(2) + 10**(-8), 1/math.sqrt(2) - 10**(-7),\
#             1/math.sqrt(2) + 10**(-7), 1/math.sqrt(2) - 10**(-6),\
#             1/math.sqrt(2) + 10**(-6), 1/math.sqrt(2) + 10**(-9), 1/math.sqrt(2) + 10**(-10)]
r_array = np.linspace(1/math.sqrt(2) - 0.001, 1/math.sqrt(2) + 0.001, 20)
phi1_arrays = [np.linspace(0.0, 0.001, 20), np.linspace(math.pi - 0.001, math.pi + 0.001, 20), np.linspace(2*math.pi - 0.001, 2*math.pi, 20, endpoint=False)]
phi3_arrays = [np.linspace(0.0, 0.001, 20), np.linspace(math.pi - 0.001, math.pi + 0.001, 20), np.linspace(2*math.pi - 0.001, 2*math.pi, 20, endpoint=False)]

for path_type in path_type_arr:
    counter = 0
    for i in range(len(phi1_arrays)):
        for k in range(len(phi3_arrays)):

            print('Iteration number ', counter)
            counter += 1
            function_testing_solution_CCpiC(r_array, phi1_arrays[i], phi3_arrays[k], path_type)
```
